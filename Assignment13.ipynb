#### Q1. Create two int-type variables, apply addition, subtraction, division, and multiplication, store the results, and print the data in a specific format.

In Python, we can create variables to store integer values and perform arithmetic operations. The results of these operations can be stored in separate variables and printed in a formatted manner.

```python
# Creating two integer variables
num1 = 20
num2 = 5

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division_float = num1 / num2  # Floating-point division
division_floor = num1 // num2  # Floor division

# Printing the results
print(f"First variable is {num1} & second variable is {num2}.")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division (float): {num1} / {num2} = {division_float}")
print(f"Division (floor): {num1} // {num2} = {division_floor}")
```

#### Q2. Difference between the operators `**` and `^`.


1. **`**` (Exponentiation Operator):**
   - This operator raises the left operand to the power of the right operand.
   - Example: `2 ** 3` calculates \(2^3 = 8\).

2. **`^` (Bitwise XOR Operator):**
   - This operator performs a bitwise XOR operation on two integers. For each bit, the result is `1` if the bits are different, and `0` if they are the same.
   - Example: `5 ^ 3` results in `6` (binary `0101 ^ 0011 = 0110`).

```python
# Example of `**`
result_exponent = 2 ** 3
print(f"2 ** 3 = {result_exponent}")

# Example of `^`
result_xor = 5 ^ 3
print(f"5 ^ 3 = {result_xor}")
```


#### Q3. List the logical operators.


Logical operators in Python are used to combine conditional statements. They operate on boolean values (`True` or `False`).

- **`and`:** Returns `True` if both conditions are `True`.
- **`or`:** Returns `True` if at least one condition is `True`.
- **`not`:** Reverses the boolean value (negation).

```python
# Examples of logical operators
x = True
y = False

# Logical AND
print(f"x and y = {x and y}")  # False

# Logical OR
print(f"x or y = {x or y}")  # True

# Logical NOT
print(f"not x = {not x}")  # False
```


#### Q4. Explain right shift operator and left shift operator with examples.


1. **Right Shift (`>>`):**
   - Moves the bits of a number to the right by the specified number of positions.
   - Equivalent to integer division by powers of 2.
   - Example: `8 >> 1` results in `4` (binary `1000 >> 1 = 0100`).

2. **Left Shift (`<<`):**
   - Moves the bits of a number to the left by the specified number of positions.
   - Equivalent to multiplication by powers of 2.
   - Example: `3 << 2` results in `12` (binary `0011 << 2 = 1100`).

```python
# Examples of right shift and left shift
num = 8
right_shift = num >> 1  # Divide by 2
left_shift = num << 2   # Multiply by 4

print(f"Right Shift: {num} >> 1 = {right_shift}")
print(f"Left Shift: {num} << 2 = {left_shift}")
```


#### Q5. Create a list of integers and check if `10` is present in the list.


We can create a list of integers and use the `in` keyword to check for the presence of a specific value.

```python
# Creating a list of integers
numbers = [1, 2, 3, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]

# Checking if 10 is present in the list
if 10 in numbers:
    print("10 is present in the list.")
else:
    print("10 is not present in the list.")
