- **Variable**: A name used to store data values, which can include strings, numbers, or other types.

 Example 
```python
string_example = "hello"  # This is a string.
variable_example = 42      # This is a variable storing an integer.
```

 

#### Q9. What three functions can be used to get the integer, floating-point number, or string version of a value? 

Functions 
1. `int()` - Converts a value to an integer.
2. `float()` - Converts a value to a floating-point number.
3. `str()` - Converts a value to a string.

Example 
```python
int_value = int("10")      # Converts string to integer
float_value = float(10)    # Converts integer to float
string_value = str(10)     # Converts integer to string
```

 

#### Q10. Why does this expression cause an error? How can you fix it? 

Problematic Code 
```python
'I have eaten ' + 99 + ' burritos.'  # TypeError
```

 
The error occurs because you cannot concatenate a string with an integer directly.

#### Solution 
Convert the integer to a string using the `str()` function:
```python
'I have eaten ' + str(99) + ' burritos.'
```

#### Output 
```
'I have eaten 99 burritos.'
```
